```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("monthly_sales.csv")
df
```

```python
df['Month, Year'] = df['Month, Year'].apply(lambda x:'-'.join([x.split('-')[0],'01','20'+x.split('-')[1]]))
```

```python
burgers = ['HM', 'CF', 'FF']
locs = ['NE','SW','NW','SE','C']
for burger in burgers:
    df[burger] = df.apply(lambda x: sum([x[burger+'-'+loc] for loc in locs]),axis=1)
```

```python
no_loc = df[['Month, Year']+burgers]
no_loc = no_loc.set_index(['Month, Year'])
no_loc
```

```python
no_loc.plot.line()
no_loc.to_csv("monthly_no_loc.csv")
```

```python
no_loc[burgers[0]].plot.line()
no_loc[[burgers[0]]].to_csv("monthly_no_loc_combine.csv")
```

```python
no_loc[burgers[1]].plot.line()
```

```python
no_loc[burgers[2]].plot.line()
```

```python
no_loc.sum(axis = 1).plot.line()
```

## Daily

```python
df = pd.read_csv('daily_sales.csv')
df
```

```python
df = df.rename({"Unnamed: 0":"Weekday"}, axis = 1)
df
```

```python
by_weekday = df.groupby(['Weekday']).mean()
by_weekday = by_weekday.loc[['Monday', 'Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']]
by_weekday
```

```python
by_weekday_total = by_weekday.sum(axis = 1)
by_weekday_total.plot.bar()
by_weekday_total.astype(int).to_csv("daily_by_weekday_total.csv")
```

```python
for burger in burgers:
    by_weekday[burger] = by_weekday.apply(lambda x: sum([x[burger+'-'+loc] for loc in locs]),axis=1)
by_weekday_no_loc = by_weekday[burgers]
by_weekday_no_loc
```

```python
by_weekday_no_loc['HM'].plot.bar()
by_weekday_no_loc.to_csv('daily_by_weekday_no_loc.csv')
```
